```python
news_titles = []
news_dates = []
news_text = []
news_link = []
```

```python
search_r1 = 'цены'
search_r2 = 'акции'
search_r3 = 'российских'
search_r4 = 'компаний'
```

```python
from selenium.webdriver.common.by import By
```

```python
driver.get('https://www.rbc.ru/search/?query=' + search_r1 + '%20'+ search_r2 +'%20'+ search_r3 +'%20'+ search_r4)

driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") # scroll to the end of the page

page_source = driver.page_source
soup = bs4.BeautifulSoup(page_source, 'lxml') 

news_titles = soup.find_all(class_='search-item__title')
news_dates = soup.find_all(class_='search-item__category')
news_text = soup.find_all(class_='search-item__text')
news_links = soup.find_all(class_="search-item__link") 
```
